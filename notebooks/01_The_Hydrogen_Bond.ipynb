We will visualize how two separate hydrogen atoms come together to form **bonding** and **anti-bonding** molecular orbitals.

### 2. Defining the Geometry

We start by placing two Hydrogen atoms at a standard bond distance of **0.74 Å**.

```python
from orbital_play import engine, show_orbital

dist = 0.74
geometry = engine.generate_geometry("H2", dist=dist)
print(f"Geometry:\n{geometry}")
```

### 3. Solving the Schrödinger Equation

We use the **Hartree-Fock** method to calculate the molecular orbitals. This process solves the equations that describe where the electrons are likely to be found.

```python
mol, mf = engine.run_calculation(geometry)
summary = engine.get_molecule_summary(mol, mf)
print(f"Total Energy: {summary['energy']:.6f} Ha")
```

### 4. Visualizing the Bonding Orbital

The first molecular orbital (MO 1) is the **bonding** orbital. Notice how the electron cloud (the isosurface) is concentrated *between* the two nuclei. This acts like "glue" that holds the atoms together.

```python
cube_data = engine.generate_cube_string(mol, mf, mo_index=0)
show_orbital(mol, cube_data, isoval=0.05)
```

### 5. Visualizing the Anti-bonding Orbital

The second molecular orbital (MO 2) is the **anti-bonding** orbital. Notice the **node** in the middle where the probability of finding an electron is zero. The phases (Blue and Red) are opposite, leading to destructive interference!

```python
cube_data_star = engine.generate_cube_string(mol, mf, mo_index=1)
show_orbital(mol, cube_data_star, isoval=0.05)
```
