# OrbitalPlay: The Art of the Pi-Bond

In previous lessons, we looked at single bonds ($\sigma$-bonds). But what happens in a double bond? In this notebook, we explore **Ethylene ($C_2H_4$)** and **Carbon Dioxide ($CO_2$)** to see the beauty of the $\pi$-bond.

```python
!pip install orbital-play pyscf py3Dmol --quiet
from orbital_play import engine, show_orbital
import numpy as np
```

## 1. Ethylene ($C_2H_4$): The Planar Double Bond

In Ethylene, each carbon is $sp^2$ hybridized. They form a single $\sigma$-bond along the C-C axis, but they also have unhybridized $p$-orbitals that overlap "side-by-side" to form a **$\pi$-bond**.

```python
geo = engine.generate_geometry("C2H4")
mol, mf = engine.run_calculation(geo)

# The Highest Occupied Molecular Orbital (HOMO) in Ethylene is the Pi-bond
# It is MO index 8 (index 7 starting from 0)
cube = engine.generate_cube_string(mol, mf, mo_index=7)
show_orbital(mol, cube, isoval=0.04)
```

Observe how the density is located **above and below** the molecular plane. This "cloud" of electrons is what makes double bonds so reactive!

## 2. Carbon Dioxide ($CO_2$): Orthogonal Pi-Systems

In $CO_2$, the carbon forms *two* $\pi$-bonds. Because of symmetry, these two bonds must be **orthogonal** (at 90 degrees) to each other.

```python
geo_co2 = engine.generate_geometry("CO2")
mol_co2, mf_co2 = engine.run_calculation(geo_co2)

# Let's look at the two degenerate Pi-orbitals
cube_pi1 = engine.generate_cube_string(mol_co2, mf_co2, mo_index=6)
cube_pi2 = engine.generate_cube_string(mol_co2, mf_co2, mo_index=7)

print("Pi-Bond 1 (X-Z plane):")
show_orbital(mol_co2, cube_pi1, isoval=0.04)

print("Pi-Bond 2 (Y-Z plane):")
show_orbital(mol_co2, cube_pi2, isoval=0.04)
```
