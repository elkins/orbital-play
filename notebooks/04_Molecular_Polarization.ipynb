# OrbitalPlay: Molecular Polarization

In this notebook, we use the **Stark Wand** to see how molecular orbitals respond to an external electric field. This is the fundamental physical basis for **Intermolecular Forces**.

```python
!pip install orbital-play pyscf py3Dmol --quiet
from orbital_play import engine, show_orbital
import py3Dmol
```

## 1. The Undisturbed Water Molecule
First, let's look at the Highest Occupied Molecular Orbital (HOMO) of Water in its natural state.

```python
geo = engine.generate_geometry("H2O")
mol, mf = engine.run_calculation(geo)
cube_orig = engine.generate_cube_string(mol, mf, mo_index=4)
show_orbital(mol, cube_orig)
```

## 2. Bringing in the Stark Wand
Now, let's place a **+1.0 charge** (a proton) nearby. Watch how the electron density is "pulled" toward the positive charge.

```python
# Place the wand 2.0 Angstroms away on the X-axis
wand_pos = (2.0, 0.0, 0.0)
mol_p, mf_p = engine.run_calculation(geo, wand_coords=wand_pos, wand_charge=1.0)

cube_p = engine.generate_cube_string(mol_p, mf_p, mo_index=4)
show_orbital(mol_p, cube_p)
```

## 3. Comparing the States
The orbitals have physically stretched! This shift in electron density creates an **induced dipole**, which is what allows neutral molecules to attract each other.
