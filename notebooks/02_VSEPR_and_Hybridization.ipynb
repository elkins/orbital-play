# OrbitalPlay: VSEPR and Hybridization

Why is water bent? Why is ammonia a pyramid? In this notebook, we explore the **Valence Shell Electron Pair Repulsion (VSEPR)** theory and the concept of **Orbital Hybridization**.

We will compare three molecules: **Methane ($CH_4$)**, **Ammonia ($NH_3$)**, and **Water ($H_2O$)**.

```python
# Setup
!pip install orbital-play pyscf py3Dmol --quiet
from orbital_play import engine, show_orbital
print("Ready!")
```

## 1. Methane ($CH_4$): The Perfect Tetrahedron

In Methane, the Carbon atom is surrounded by 4 bonding pairs of electrons. These pairs repel each other equally, resulting in a perfect tetrahedral geometry with angles of **109.5°**.

Let's visualize the bonding orbitals.

```python
geo_ch4 = engine.generate_geometry("CH4", dist=1.09)
mol, mf = engine.run_calculation(geo_ch4)

# Visualize one of the bonding orbitals (MO 2, 3, 4, or 5)
cube = engine.generate_cube_string(mol, mf, mo_index=1)
show_orbital(mol, cube, isoval=0.05)
```

## 2. Ammonia ($NH_3$): The Lone Pair Effect

Ammonia has 3 bonding pairs and **1 lone pair**. According to VSEPR theory, lone pairs take up more space and repel bonding pairs more strongly than bonding pairs repel each other.

This 

pushes the Hydrogen atoms closer together, reducing the angle from 109.5° to approximately **107°**.

```python
geo_nh3 = engine.generate_geometry("NH3", dist=1.01, angle=106.7)
mol, mf = engine.run_calculation(geo_nh3)

# MO 5 in NH3 corresponds to the Lone Pair
cube = engine.generate_cube_string(mol, mf, mo_index=4)
show_orbital(mol, cube, isoval=0.05)
```

## 3. Water ($H_2O$): Two Lone Pairs

Water has 2 bonding pairs and **2 lone pairs**. The repulsion from two lone pairs is even stronger, pushing the O-H bonds even closer to an angle of **104.5°**.

```python
geo_h2o = engine.generate_geometry("H2O", dist=0.96, angle=104.5)
mol, mf = engine.run_calculation(geo_h2o)

# MO 4 and 5 in H2O represent the lone pairs
cube = engine.generate_cube_string(mol, mf, mo_index=4)
show_orbital(mol, cube, isoval=0.05)
```
