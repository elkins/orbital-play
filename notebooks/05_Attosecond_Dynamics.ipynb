# OrbitalPlay: Attosecond Dynamics

The **2023 Nobel Prize in Physics** was awarded for experiments that allow us to see electrons moving in real-time. In this notebook, we visualize how electron density "sloshes" when a molecule is excited by light.

```python
!pip install orbital-play pyscf py3Dmol --quiet
from orbital_play import engine
import py3Dmol
```

## 1. Capturing the Motion
We will simulate a coherent superposition of the ground state and the first excited state of **Hydrogen ($H_2$)**.

```python
geo = engine.generate_geometry("H2", dist=0.74)
mol, mf = engine.run_calculation(geo)

# Generate 10 frames of the electron 'sloshing'
frames = engine.get_attosecond_frames(mol, mf, n_frames=10, amplitude=0.8)
```

## 2. Interactive Movie
We can't easily show a video in a static cell, but we can look at the two extreme ends of the "slosh" (the left and right peak of the oscillation).

```python
from orbital_play import show_orbital

print("Frame 1: Density shifts LEFT")
show_orbital(mol, frames[0], isoval=0.08)

print("Frame 5: Density shifts RIGHT")
show_orbital(mol, frames[5], isoval=0.08)
```

This oscillation happens on the **Attosecond** timescale ($10^{-18}$ seconds)—so fast that it takes a Nobel-winning laser to see it!
